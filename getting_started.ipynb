```python
import requests
```

### Loading API KEY from text file

```python
with open('api_key.txt') as file:
    api_key = file.read()
api_key
```

### Setting Regional Endpoint and Summoner Name

- The host it's different for each server

```python
host = 'br1.api.riotgames.com'
summoner = 'ehtu0uteuirmao'
```

#### Getting Summoner By Summoner Name

```python
r = requests.get('https://'+host+'/lol/summoner/v4/summoners/by-name/'+summoner+'?api_key='+api_key)
```

```python
r.json()
```

#### Getting Summoner ID

```python
summ_id = r.json()['id']
summ_id
```

#### Getting Account ID

```python
acc_id = r.json()['accountId']
acc_id
```

### CHAMPION-MASTERY-V4

```python
champ_mastery = '/lol/champion-mastery/v4/champion-masteries/by-summoner/'+summ_id
```

```python
result_request = requests.get('https://'+host+champ_mastery+'?api_key='+api_key)
result_request
```

```python
result_request.json()
```

### Match List

```python
match_list_result = requests.get('https://'+host+'/lol/match/v4/matchlists/by-account/'+acc_id+'?api_key='+api_key)
match_list_result.json()
```

```python
match_list_result.json()['matches'][0]
```

```python
match_list_result.json()['matches'][0]['gameId']
```

### Match by Match

```python
matchId = match_list_result.json()['matches'][0]['gameId']
```

```python
by_match_result = requests.get('https://'+host+'/lol/match/v4/matches/'+str(matchId)+'?api_key='+api_key)
by_match_result.json()
```

### Match Timeline

```python
matchId = match_list_result.json()['matches'][0]['gameId']
```

```python
match_timeline_result = requests.get('https://'+host+'/lol/match/v4/timelines/by-match/'+str(matchId)+'?api_key='+api_key)
match_timeline_result.json()
```
